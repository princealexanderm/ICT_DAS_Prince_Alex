###Fill missing values in salary currency with employee residence countries currency that is if employee residence is UK, fill it with GBP, if US fill it with us dolars, if DE, fill it with EUR, if IN fill it with INR,  if Jp fill it with JPY. If employee residence is MX or CN fill it randomly with randomly with GBP, US, EUR, JPY, INR

```python
# Fill missing 'salary_currency' based on 'employee_residence'
def fill_salary_currency(row):
    if pd.isnull(row['salary_currency']):
        if row['employee_residence'] == 'UK':
            return 'GBP'
        elif row['employee_residence'] == 'US':
            return 'us dolars'
        elif row['employee_residence'] == 'DE':
            return 'EUR'
        elif row['employee_residence'] == 'IN':
            return 'INR'
        elif row['employee_residence'] == 'JP':
            return 'JPY'
        elif row['employee_residence'] in ['MX', 'CN']:
            return np.random.choice(['GBP', 'us dolars', 'EUR', 'JPY', 'INR'])
        else:
            return np.nan  # Or another strategy for other countries
    return row['salary_currency']

dff['salary_currency'] = dff.apply(fill_salary_currency, axis=1)
```

##app.py

```python
import pandas as pd
from flask import Flask,render_template,request
import pickle
import numpy as np



app = Flask(__name__)

file_name="dsa_job.csv"

df=pd.read_csv(file_name)

rf_model = pickle.load(open('model.pkl', 'rb'))
encoder = pickle.load(open('encoder.pkl', 'rb'))

job_title_values=df['job_title'].unique()
job_category_values=df['job_category'].unique()
experience_level_values=df['experience_level'].unique()
employment_type_values=df['employment_type'].unique()
work_setting_values=df['work_setting'].unique()
company_size_values=df['company_size'].unique()

@app.route('/')
def home():
    return render_template('first.html',job_title_values=job_title_values,job_category_values=job_category_values,
                           experience_level_values=experience_level_values,employment_type_values=employment_type_values,
                           work_setting_values=work_setting_values,company_size_values=company_size_values)

@app.route('/predict',methods=['POST'])
def predict():
    #Categorical
    job_title = request.form['job_title']
    job_category = request.form['job_category']
    experience_level = request.form['experience_level']
    employment_type = request.form['employment_type']
    work_setting = request.form['work_setting']
    company_size = request.form['company_size']

    encoded_df=pd.DataFrame(encoder.transform([[job_title,job_category,experience_level,employment_type,work_setting,company_size]]), columns=encoder.get_feature_names_out())

    #input_features = np.array([[volatile_acidity,citric_acid,residual_sugar,chlorides,total_sulfur_dioxide,density,pH,sulphates,alcohol]])
    pred = rf_model.predict(encoded_df)

    target = pred[0]
    #species = wine.target_names[target]
    return render_template('first.html',pred_result = target)

if __name__=='__main__':
    app.run(debug=True)
```
