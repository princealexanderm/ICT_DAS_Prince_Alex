```python
import pandas as pd
df2=pd.read_csv('/content/Sales_add.csv')
df2
```

**1.** The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

**T-Test**

image.png

```python
mean_before=df2['Sales_before_digital_add(in $)'].mean()
mean_before
```

```python
mean_after=df2['Sales_After_digital_add(in $)'].mean()
mean_after
```

```python
std_after=df2['Sales_After_digital_add(in $)'].std()
std_after
```

```python
sample_mean=231123.72727272726
population_mean=149239.95454545456
std_dev=25556.777060725042
sample_size=21
alpha=0.05
```

```python
import math
std_error=std_dev/math.sqrt(sample_size)
std_error
```

```python
# t value calculation
t_value=(sample_mean-population_mean)/std_error
t_value
```

image.png

```python
t_critical=1.75 # from t table for degree of freedom 20 which is 21-1
```

```python
if t_value>t_critical:
  print('T-score:{t_value:.2f}')
  print('reject the null hypothesis,there is an increase in sales after stepping into digital marketing')
else:
  print('T-score:{t_value:.2f}')
  print('reject the alternate, not enough evidence to support the idea that there is an increase in sales after stepping into digital marketing')
```

**Digital marketing improved sales**
